```python
import pandas as pd

# Load the cleaned and merged dataset
movies_df = pd.read_csv('merged_cleaned_tmdb_5000_movies.csv')

# Check for duplicate ids
duplicate_ids = movies_df[movies_df.duplicated(['id'], keep=False)]
if not duplicate_ids.empty:
    print("Duplicate IDs found in the dataset:")
    print(duplicate_ids)
else:
    print("No duplicate IDs found.")
```

```python
# Drop duplicates, keeping the first occurrence
movies_df = movies_df.drop_duplicates(subset=['id'], keep='first')

# Save the cleaned dataset back to CSV
movies_df.to_csv('cleaned_merged_tmdb_5000_movies.csv', index=False)
```
